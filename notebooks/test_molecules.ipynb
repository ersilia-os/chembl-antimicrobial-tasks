```python
import pandas as pd
import os

PATH = "/home/acomajuncosa/Documents/chembl-antimicrobial-tasks_old/output"
OUTPATH = "/home/acomajuncosa/Desktop"
PATHOGENS = ['pfalciparum', 'mtuberculosis', 'kpneumoniae', 'abaumannii', 'saureus', 'paeruginosa']
```

```python
for pathogen in PATHOGENS:

    print(pathogen)

    # Load REF SET
    ref_set = pd.read_csv(os.path.join(PATH, f"{pathogen}_organism", "018_ref_set.csv"))[:1000]

    # Load mappings
    mappings = pd.read_csv(os.path.join(PATH, f"{pathogen}_organism", f"012_{pathogen}_cleaned.csv"))
    mappings = {i:j for i,j in zip(mappings['inchikey'], mappings['smiles'])}

    # Map
    ref_set['smiles'] = [mappings[i] for i in ref_set['inchikey']]

    # Save
    ref_set.to_csv(os.path.join(OUTPATH, f"{pathogen}_top1000.csv"), index=False)
```
