```python
import pandas as pd
import pickle
import os
```

```python
root = "."
pathogen_code = 'mtuberculosis'

ASSAY_DATA_RANGES = pd.read_csv(os.path.join(root, "..", "output", pathogen_code, "assay_data_ranges.csv"))

ASSAY_DATA_RANGES = ASSAY_DATA_RANGES[(ASSAY_DATA_RANGES['cpds'] > 100) & 
                                      (ASSAY_DATA_RANGES['ratio'] < 0.2) & 
                                      (ASSAY_DATA_RANGES['ratio'] > 0.001)].reset_index(drop=True)
```

```python
models_to_train = ASSAY_DATA_RANGES[['assay_chembl_id', 'activity_type', 'unit']].values
pickle.dump(models_to_train, open(os.path.join(root, "..", "tmp", "models_to_train.pkl"), "wb"))
```
